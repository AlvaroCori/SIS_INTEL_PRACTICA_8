```python
import pomegranate as pg

 

smokeD = pg.DiscreteDistribution({'yes': 0.25, 'no': 0.75})

covidD = pg.DiscreteDistribution({'yes': 0.1, 'no': 0.9})

hospitalD = pg.ConditionalProbabilityTable(

    [['yes', 'yes', 'yes', 0.9], ['yes', 'yes', 'no', 0.1],

['yes', 'no', 'yes', 0.1], ['yes', 'no', 'no', 0.9],['no', 'yes', 'yes', 0.9], ['no', 'yes', 'no', 0.1],

     ['no', 'no', 'yes', 0.01], ['no', 'no', 'no', 0.99]],

    [smokeD, covidD])

 

smoke = pg.Node(smokeD, name="smokeD")

covid = pg.Node(covidD, name="covidD")

hospital = pg.Node(hospitalD, name="hospitalD")

 

model = pg.BayesianNetwork("Covid Collider")

model.add_states(smoke, covid, hospital)

model.add_edge(smoke, hospital)

model.add_edge(covid, hospital)

model.bake()

#You could then calculate P(covid|smoking, hospital) = 0.5 with

model.predict_proba({'smokeD': 'yes', 'hospitalD': 'yes'})

#and P(covid|Â¬smoking, hospital)=0.91 with

model.predict_proba({'smokeD': 'no', 'hospitalD': 'yes'})
```

```python
import numpy as np
import pandas as pd
import pymc3 as pm
import seaborn as sns
import theano
import warnings
from numpy.random import binomial, randn, uniform
from sklearn.model_selection import train_test_split
```

```python
%matplotlib inline

sns.set()
warnings.filterwarnings('ignore')
```

```python
alpha_0 = 1
alpha_1 = 1.25

beta_0 = 1
beta_1 = 1.25

sigma = 0.75
```

```python
n_samples = 1000

category = binomial(n=1, p=0.5, size=n_samples)
x = uniform(low=0, high=10, size=n_samples)

y = ((1 - category) * alpha_0 + category * alpha_1
     + ((1 - category) * beta_0 + category * beta_1) * x
     + sigma * randn(n_samples))

model_data = pd.DataFrame({'y': y, 'x': x, 'category': category})

display(model_data.head())
_ = sns.relplot(x='x', y='y', hue='category', data=model_data)
```

```python
train, test = train_test_split(
    model_data, test_size=0.2, stratify=model_data.category)
```

```python
y_tensor = theano.shared(train.y.values.astype('float64'))
x_tensor = theano.shared(train.x.values.astype('float64'))
cat_tensor = theano.shared(train.category.values.astype('int64'))
```

```python
with pm.Model() as model:
    alpha_prior = pm.HalfNormal('alpha', sd=2, shape=2)
    beta_prior = pm.Normal('beta', mu=0, sd=2, shape=2)
    sigma_prior = pm.HalfNormal('sigma', sd=2, shape=1)
    mu_likelihood = alpha_prior[cat_tensor] + beta_prior[cat_tensor] * x_tensor
    y_likelihood = pm.Normal('y', mu=mu_likelihood, sd=sigma_prior, observed=y_tensor)
```

```python
with model:
    hmc_trace = pm.sample(draws=50, tune=10, cores=2)
```

```python
pm.traceplot(hmc_trace)
pm.summary(hmc_trace)
```
